```python
import os

import pandas as pd
```

## Nextflow preparation

```python
sra_metadata = pd.read_csv(os.path.join(INPUT_DIR, "nfcore_rnaseq_3.14.0", "sra_explorer_metadata.tsv"), sep="\t")
sra_metadata
```

```python
DATA_DIR = "/n/scratch/users/w/wit498/bed_rest_bulkRNAseq/input"
samples = []

for sample_sra, sample in sra_metadata.groupby("Accession"):
    
    sample_name = sample["FastQ nice filename"].iloc[0].split(".")[0].rsplit("_", 1)[0]
    fastq1_fp = os.path.join(DATA_DIR, f"{sample_sra}_1.fastq.gz")
    fastq2_fp = os.path.join(DATA_DIR, f"{sample_sra}_2.fastq.gz")
    
    samples.append(
        {
            "sample": sample_name,
            "fastq_1": fastq1_fp,
            "fastq_2": fastq2_fp,
            "strandedness": "auto"
            
        }
    )

samples_df = pd.DataFrame(samples)

samples_df
```

```python
samples_df.to_csv(os.path.join(INPUT_DIR, "samplesheet.csv"), index=False)
```

## Kallisto tximport

```python
import glob
```

```python
sample_fps = glob.glob(os.path.join(INPUT_DIR, "nfcore_rnaseq_3.14.0", "kallisto", "*", "abundance.h5"))
samples = pd.DataFrame({"filepath": sample_fps, })
samples[["patient", "tmp"]] = samples.filepath.str.rsplit("_", expand=True, n=5).iloc[:,[1, 2]]
samples
```

```python
samples.loc[samples.tmp == "POST", "tmp"] = "Post"
samples.loc[samples.tmp == "PRE", "tmp"] = "Pre"
```

```python
sample_annot = pd.read_csv(os.path.join(INPUT_DIR, "samples.csv"))
sample_annot
```

```python
samples = samples.set_index(["patient", "tmp"]).join(sample_annot.set_index(["patient", "tmp"])).reset_index()
samples["sample_name"] = samples.patient + "_" + samples.tmp
samples
```

```python
samples.to_csv(os.path.join(INPUT_DIR, "samples_tximport.csv"), index=False)
```
